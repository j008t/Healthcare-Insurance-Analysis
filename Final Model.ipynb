```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.cbook import boxplot_stats  
import statsmodels.api as sm
from sklearn.model_selection import train_test_split,GridSearchCV, cross_val_score, cross_val_predict
from statsmodels.stats.outliers_influence import variance_inflation_factor 
from sklearn.tree import DecisionTreeRegressor
from sklearn import ensemble
import numpy as np
import pickle
```

```python
#Visualizing age column with a histogram
fig,axes=plt.subplots(1,2,figsize=(10,5))
sns.histplot( data['age'] , color="skyblue",ax=axes[0])
sns.histplot( data['bmi'] , color="olive",ax=axes[1])
plt.show()
```

```python
#Visualizing age column with a boxplot
fig,axes=plt.subplots(1,2,figsize=(10,5))

sns.boxplot(x = 'age', data = data, ax=axes[0])
sns.boxplot(x = 'bmi', data = data, ax=axes[1])

plt.show()
```

```python
#Finding the outlier values in the bmi column
outlier_list = boxplot_stats(data.bmi).pop(0)['fliers'].tolist()
print(outlier_list)

#Finding the number of rows containing outliers
outlier_bmi_rows = data[data.bmi.isin(outlier_list)].shape[0]
print("Number of rows contaning outliers in bmi : ", outlier_bmi_rows)

#Percentage of rows which are outliers
percent_bmi_outlier = (outlier_bmi_rows/data.shape[0])*100
print("Percentage of outliers in bmi columns : ", percent_bmi_outlier)
```

```python
#Converting age into age brackets
print("Minimum value for age : ", data['age'].min(),"\nMaximum value for age : ", data['age'].max())

#Age between 18 to 40 years will fall under young
#Age between 41 to 58 years will fall under mid-age
#Age above 58 years will fall under old age

data.loc[(data['age'] >=18) & (data['age'] <= 40), 'age_group'] = 'young'
data.loc[(data['age'] >= 41) & (data['age'] <= 58), 'age_group'] = 'mid-age'
data.loc[data['age'] > 58, 'age_group'] = 'old'
```

```python
#Removing outier from bmi column
data_clean = data[~data.bmi.isin(outlier_list)]
sns.boxplot(x = 'bmi', data = data_clean)
```

```python
fig,axes=plt.subplots(1,5,figsize=(20,8))
sns.countplot(x = 'sex', data = data_clean, palette = 'magma',ax=axes[0])
sns.countplot(x = 'children', data = data_clean, palette = 'magma',ax=axes[1])
sns.countplot(x = 'smoker', data = data_clean, palette = 'magma',ax=axes[2])
sns.countplot(x = 'region', data = data_clean, palette = 'magma',ax=axes[3])
sns.countplot(x = 'age_group', data = data_clean, palette = 'magma',ax=axes[4])
```
